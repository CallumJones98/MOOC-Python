```python
# The longest string 5.1 

def longest(strings: list):
    
    longest_string = strings[0]
    
    for word in strings:
        if len(word) >= len(longest_string):
            longest_string = word
        else:
            continue

    return longest_string

longest(["hi", "hiya", "hello", "howdydoody", "hi there"])
```

```python
# Factorials 

def factorials(n: int):
    results = {}
    run = 1
    
    for val in range(1, n + 1):
        run *= val  
        results[val] = run
        
    return results

k = factorials(5)

print(k[5])
```

```python
# The oldest person

def oldest_person(people: list):

    max_age = 2000
    chosen_person = None

    for person in people:
        name, yob = person
        if yob <= max_age:
            max_age = yob
            chosen_person = name
        else:
            continue

    return chosen_person
        
p1 = ("Adam", 1977)
p2 = ("Ellen", 1985)
p3 = ("Mary", 1953)
p4 = ("Ernest", 1997)
people = [p1, p2, p3, p4]

print(oldest_person(people))
```

```python
# Student Database 

students = {}

def add_student(database: dict, student: str):
    
    new_student = {student: []}
    database.update(new_student)


def print_student(database: dict, student: str):

    if student not in database:
        print(f'{student}: no such person in the database')
    elif student in database:
        if len(database[student]) < 1:
                print(f'{student}:')
                print(f'  no completed course')
        else:
            print(f'{student}:')
            print(f' {len(database[student])} complete courses:')
            count = 0
            for course_name, grade in database[student]:
                print(f'  {course_name} {grade}')
                count += grade
            print(f' average grade {count / len(database[student])}')


def add_course(database: dict, student: str, course: tuple):
    
    if student not in database:
        return 'Student not in database'

    course_name, new_grade = course

    if new_grade < 1:
        return 

    # Check if course already exists
    course_found = False
    
    for index, (existing_course, existing_grade) in enumerate(database[student]):
        if existing_course == course_name:
            course_found = True
            if new_grade > existing_grade:
                database[student][index] = (course_name, new_grade)
            break 
    
    if not course_found:
        database[student].append((course_name, new_grade))


def summary(database: dict):

    # students 
    print(f'students {len(database)}')

    #most courses completed 

    max_len = 0 
    chosen_student = None
    for key, value in db.items():
        if len(value) >= max_len:
            max_len = len(value)
            chosen_student = key
        else:
            continue
            
    print(f'most courses completed {max_len} {chosen_student}')

    # best average grade

    averages = {name: sum(score for course, score in courses) / len(courses) for name, courses in database.items()}

    print(f'best average grade {max(averages.values())} {max(averages, key=averages.get)}')
    
    
                    
    
        
students = {}
add_student(students, "Peter")
add_student(students, "Eliza")
add_course(students, "Peter", ("Data Structures and Algorithms", 1))
add_course(students, "Peter", ("Introduction to Programming", 1))
add_course(students, "Peter", ("Advanced Course in Programming", 1))
add_course(students, "Eliza", ("Introduction to Programming", 5))
add_course(students, "Eliza", ("Introduction to Computer Science", 4))
summary(students)
```
