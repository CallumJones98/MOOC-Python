```python
# Matrix

def matrix_sum():
  with open('matrix.txt') as new_file:
      total_li = []
      for line in new_file:
          line = line.replace("\n", "")
          nums = line.split(',')
          for num in nums:
              total_li.append(int(num))

      return sum(total_li)

def matrix_max():
  with open('matrix.txt') as new_file:
      total_li = []
      for line in new_file:
          line = line.replace("\n", "")
          nums = line.split(',')
          for num in nums:
              total_li.append(int(num))

      return max(total_li)
  
def row_sums():
  with open('matrix.txt') as new_file:
      row_sums_total = []
      for line in new_file:
          line = line.replace("\n", "")
          nums = line.split(',')
          val = []
          for num in nums:
              val.append(int(num))
          row_sums_total.append(sum(val))
      return row_sums_total

```

```python
# Recipe Search

def search_by_name(filename: str, word: str):
    recipe_dict = {}

    with open(filename, 'r') as file:
        lines = [line.strip() for line in file]

    recipe_name = None
    ingredients = []
    cooking_time = None

    for line in lines + [""]:  # add a sentinel blank at the end
        if not line:
            if recipe_name:  # save the current recipe
                recipe_dict[recipe_name] = [cooking_time] + ingredients
                recipe_name = None
                ingredients = []
                cooking_time = None
            continue

        if recipe_name is None:
            recipe_name = line
        elif cooking_time is None:
            cooking_time = int(line)  
        else:
            ingredients.append(line)

    return [name for name in recipe_dict if word.lower() in name.lower()]


def search_by_time(filename: str, prep_time: int):
    recipe_dict = {}

    with open(filename, 'r') as file:
        # State for the current recipe
        recipe_name = None
        cooking_time = None
        ingredients = []

        # Add a sentinel blank line so the last recipe gets saved
        for raw in list(file) + ["\n"]:
            line = raw.strip()

            if not line:
                # end of current recipe → save it (if we have one)
                if recipe_name is not None:
                    recipe_dict[recipe_name] = [cooking_time] + ingredients
                    recipe_name = None
                    cooking_time = None
                    ingredients = []
                continue

            if recipe_name is None:
                recipe_name = line
            elif cooking_time is None:
                cooking_time = int(line)   # only the line right after name
            else:
                ingredients.append(line)

    li_of_li = []
    for key, val in recipe_dict.items():
        if val[0] < prep_time:  # keep your original '<' semantics
            li_of_li.append(f'{key}, preparation time {val[0]} min')

    return li_of_li


def search_by_ingredient(filename: str, ingredient: str):

    recipe_dict = {}
    
    with open(filename, 'r') as file:
        # State for the current recipe
        recipe_name = None
        cooking_time = None
        ingredients = []

        # Add a sentinel blank line so the last recipe gets saved
        for raw in list(file) + ["\n"]:
            line = raw.strip()

            if not line:
                # end of current recipe → save it (if we have one)
                if recipe_name is not None:
                    recipe_dict[recipe_name] = [cooking_time] + ingredients
                    recipe_name = None
                    cooking_time = None
                    ingredients = []
                continue

            if recipe_name is None:
                recipe_name = line
            elif cooking_time is None:
                cooking_time = int(line)   # only the line right after name
            else:
                ingredients.append(line)

    ingredient_li = []

    for key, val in recipe_dict.items():
        for v in val:
            if v == ingredient:
                ingredient_li.append(f'{key}, preparation time {val[0]} min')

    return ingredient_li
    
```

```python
# City Bikes 

import csv
import math 

def get_station_data(filename: dict):

    my_dict = {}

    with open(filename) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=';')
        next(csv_reader)
        for row in csv_reader:
            my_dict[row[3]] = (float(row[0]), float(row[1]))

    return my_dict


def distance(stations: dict, station1: str, station2: str):

    long1, lat1 = stations[station1]
    long2, lat2 = stations[station2]

    x_km = (long1 - long2) * 55.26
    y_km = (lat1 - lat2) * 111.2
    distance_km = math.sqrt(x_km**2 + y_km**2)

    return distance_km

            
def greatest_distance_from_one_station(stations: dict, stations_list: list, station: str):
    start_index = stations_list.index(station)+1
    compare_list = stations_list[start_index:]
    greatest_tuple = None
    greatest = 0
    for compare in compare_list:
        d = distance(stations, station, compare)
        if d > greatest:
            greatest_tuple = (station, compare, d)
            greatest = d
    return greatest_tuple

def greatest_distance(stations: dict):
    stations_list = []
    for key, value in stations.items():
        stations_list.append(key)

    greatest_tuple = None
    greatest = 0
    for index, station in enumerate(stations_list):
        if index < len(stations_list)-1:
            stations_distance = greatest_distance_from_one_station(stations, stations_list, station)
            station1, station2, distance = stations_distance
            if distance > greatest:
                greatest_tuple = stations_distance
                greatest = distance
    
    return greatest_tuple
```

```python
# Course grading - part 4

# open file
def read_file(file_name):
    file_name = file_name.lower()
    with open(file_name) as file:
        if 'students' in file_name or 'exercises' in file_name or 'exam_points' in file_name:
            dict = {}
            for line in file:
                line = line.replace('\n', '')
                line_as_list = line.split(';')
                if line_as_list[0] == 'id':
                    continue
                dict[line_as_list[0]] = line_as_list[1:]
            return dict
        elif 'course' in file_name:
            course_info = []
            for line in file:
                line = line.replace('\n', '')
                line_as_list = line.split(':')
                course_info.append(line_as_list[1].strip())
            return course_info

def convert_dict_values_list_str_to_int(dict):
    for key, value in dict.items():
        for list_index, list_value in enumerate(value):
            value[list_index] = int(list_value)

def points_to_grade(points):
    grade = -1
    if points <= 14:
        return 0
    elif points <= 17:
        return 1
    elif points <= 20:
        return 2
    elif points <= 23:
        return 3
    elif points <= 27:
        return 4
    else:
        return 5

def write_results(students, exercises, exams, course, results_file_name):
    label_name = 'name'
    exec_nbr = 'exec_nbr'
    exec_pts = 'exec_pts.'
    exm_pts = 'exm_pts.'
    tot_pts = 'tot_pts.'
    grade = 'grade'

    students_grade = []
    report = ''
    report += (f'Introduction to Programming, 5 credits') + '\n'
    report += ('======================================') + '\n'
    report += (f'{label_name:30}{exec_nbr:10}{exec_pts:10}{exm_pts:10}{tot_pts:10}{grade:10}') + '\n'
    for id, name in students.items():
        exercise_total = sum(exercises[id])
        exercise_points = int(((exercise_total/40)*100)//10)
        exam_points = sum(exams[id])
        total_points = exercise_points+exam_points
        grade = points_to_grade(total_points)
        full_name = name[0] + ' ' + name[1]
        students_grade.append([str(id), full_name, str(grade)])
        report += (f'{full_name:30}{exercise_total:<10}{exercise_points:<10}{exam_points:<10}{total_points:<10}{grade:<10}') + '\n'

    results_txt_file_name = results_file_name + '.txt'
    results_csv_file_name = results_file_name + '.csv'
    with open(results_txt_file_name, 'w') as results_txt_file:
        results_txt_file.write(report)
    with open(results_csv_file_name, 'w') as results_csv_file:
        for record in students_grade:
            line = ';'.join(record) + '\n'
            results_csv_file.write(line)
    print(f'Results written to files {results_txt_file_name} and {results_csv_file_name}')

def main():
    if True:
        student_file = input('Student information: ')
        exercises_file = input('Exercises completed: ')
        exam_file = input('Exam points: ')
        course_file = input('Course information: ')
    else:
        student_file = 'students1.csv'
        exercises_file = 'exercises1.csv'
        exam_file = 'exam_points1.csv'
        course_file = 'course1.txt'
    
    results_file = 'results'
    
    students = read_file(student_file)
    exercises = read_file(exercises_file)
    exams = read_file(exam_file)
    course = read_file(course_file)

    convert_dict_values_list_str_to_int(exercises)
    convert_dict_values_list_str_to_int(exams)

    write_results(students, exercises, exams, course, results_file)
```

```python
# Word Search

def find_words(search_term: str):
  li = []

  with open('words.txt', 'r') as file:
      for line in file:
          clean_line = line.rstrip('\n')
          if len(search_term) == len(clean_line) and '.' in search_term:
              for search_letter, word_letter in zip(search_term, clean_line):
                  if search_letter == '.' or search_letter == word_letter:
                      continue  
                  else:
                      break     

              else:
                  li.append(clean_line)

          elif search_term.endswith('*'):
              split = search_term.index('*')
              if clean_line[0:split] == search_term[0:split]:
                  li.append(clean_line)
              else:
                  continue

          elif search_term.startswith('*'):
              if clean_line.endswith(search_term[1:]):
                  li.append(clean_line)
              else:
                  continue

          elif ('.' not in search_term) and ('*' not in search_term):
              if search_term == clean_line:
                  li.append(clean_line)
              else:
                  continue 


      return li
        
```
