```python
# Smallest Average

def smallest_average(person1: dict, person2: dict, person3: dict):
  
  person_one_avg = 0
  person_one_len = 0
  person_two_avg = 0
  person_two_len = 0
  person_three_avg = 0
  person_three_len = 0
  
  for key,value in person1.items():
    if isinstance(value, int):
      person_one_avg += value
      person_one_len += 1 
    else:
      continue

  for key,value in person2.items():
    if isinstance(value, int):
      person_two_avg += value
      person_two_len += 1 
    else:
      continue
    
  for key,value in person3.items():
    if isinstance(value, int):
      person_three_avg += value
      person_three_len += 1 
    else:
      continue
  
  
  person_one = person_one_avg / person_one_len
  person_two = person_two_avg / person_two_len
  person_three = person_three_avg / person_three_len
  smallest = min(person_one, person_two, person_three)
  
  if smallest == person_one:
    return person1
  elif smallest == person_two:
    return person2
  elif smallest == person_three:
    return person3
  else:
    return None
      
```

```python
# Shopping List 


class ShoppingList:
    def __init__(self):
        self.products = []

    def number_of_items(self):
        return len(self.products)

    def add(self, product: str, number: int):
        self.products.append((product, number))

    def item(self, n: int):
        return self.products[n - 1][0]

    def amount(self, n: int):
        return self.products[n - 1][1]


    
def total_units(my_list : list):
    unit = 0
    for i in range(1, my_list.number_of_items()+1):
        unit += my_list.amount(i)

    return unit

        
```

```python
# books of genre

class Book:
    def __init__(self, name: str, author: str, genre: str, year: int):
        self.name = name
        self.author = author
        self.genre = genre 
        self.year = year

    ##STUB:# This enables easy printing of a Book object
    def __repr__(self):
        return f"{self.name} ({self.author}), {self.year} - genre: {self.genre}"


def books_of_genre(books: list, genre: str):

    genre_books = [book for book in books if book.genre == genre]

    return genre_books


python = Book("Fluent Python", "Luciano Ramalho", "programming", 2015)
everest = Book("High Adventure", "Edmund Hillary", "autobiography", 1956)
norma = Book("Norma", "Sofi Oksanen", "crime", 2015)

books = [python, everest, norma, Book("The Snowman", "Jo Nesbø", "crime", 2007)]

print("Books in the crime genre:")
books_of_genre(books, "crime")
for book in books_of_genre(books, "crime"):
    print(f"{book.author}: {book.name}")
```

```python
#Example of methods in classes

class BankAccount:

    def __init__(self, account_number: str, owner: str, balance: float, annual_interest: float):
        self.account_number = account_number
        self.owner = owner
        self.balance = balance
        self.annual_interest = annual_interest

    # This method adds the annual interest to the balance of the account
    def add_interest(self):
        self.balance += self.balance * self.annual_interest

    # This method "withdraws" money from the account
    # If the withdrawal is successful the method returns True, and False otherwise
    def withdraw(self, amount: float):
        if amount <= self.balance:
            self.balance -= amount
            return True

        return False
```

```python
# Decreasing Counter 

class DecreasingCounter:
    def __init__(self, initial_value: int):
        self.value = initial_value
        self.base_value = initial_value

    def print_value(self):
        print("value:", self.value)

    def decrease(self):
        if self.value >= 1:
            self.value -= 1

    def set_to_zero(self):
        self.value = 0

    def reset_original_value(self):
        self.value = self.base_value
        

counter = DecreasingCounter(55)
counter.decrease()
counter.decrease()
counter.decrease()
counter.decrease()
counter.print_value()
counter.reset_original_value()
counter.print_value()
```

```python
# Statistics on numbers 

# Create NumberStats class

class NumberStats:

    def __init__(self):
        self.numbers = 0
        self.counter = 0

    def add_number(self, number:int):
        self.numbers += number
        self.counter += 1

    def count_numbers(self):
        return self.counter

    def get_sum(self):
        if self.counter != 0:
            return self.numbers
        else:
            return 0

    def average(self):
        if self.counter != 0:
            return self.numbers / self.counter
        else:
            return 0 

stats = NumberStats()
stats.add_number(3)
stats.add_number(5)
stats.add_number(1)
stats.add_number(2)
print("Numbers added:", stats.count_numbers())
print("Sum of numbers:", stats.get_sum())
print("Mean of numbers:", stats.average())
```

```python
# Part 4 - Multiple sums 

class NumberStats:

    def __init__(self):
        self.numbers = 0
        self.counter = 0

    def add_number(self, number:int):
        self.numbers += number
        self.counter += 1

    def count_numbers(self):
        return self.counter

    def get_sum(self):
        if self.counter != 0:
            return self.numbers
        else:
            return 0

    def average(self):
        if self.counter != 0:
            return self.numbers / self.counter
        else:
            return 0 

#initialise main, odd, even
values = NumberStats()
odds = NumberStats()
evens = NumberStats()

print('Please type in integer numbers: ')
user_inp = int(input(' '))

while user_inp != -1:
    values.add_number(user_inp)
    if user_inp % 2 == 0:
        evens.add_number(user_inp)
    else:
        odds.add_number(user_inp)
    
    user_inp = int(input(' '))

print(f'Sum of numbers: {values.get_sum()}')
print(f'Mean of numbers: {values.average()}')
print(f'Sum of even numbers: {evens.get_sum()}')
print(f'Sum of odd numbers: {odds.get_sum()}')


```
