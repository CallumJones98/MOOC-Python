```python
# Smallest Average

def smallest_average(person1: dict, person2: dict, person3: dict):
  
  person_one_avg = 0
  person_one_len = 0
  person_two_avg = 0
  person_two_len = 0
  person_three_avg = 0
  person_three_len = 0
  
  for key,value in person1.items():
    if isinstance(value, int):
      person_one_avg += value
      person_one_len += 1 
    else:
      continue

  for key,value in person2.items():
    if isinstance(value, int):
      person_two_avg += value
      person_two_len += 1 
    else:
      continue
    
  for key,value in person3.items():
    if isinstance(value, int):
      person_three_avg += value
      person_three_len += 1 
    else:
      continue
  
  
  person_one = person_one_avg / person_one_len
  person_two = person_two_avg / person_two_len
  person_three = person_three_avg / person_three_len
  smallest = min(person_one, person_two, person_three)
  
  if smallest == person_one:
    return person1
  elif smallest == person_two:
    return person2
  elif smallest == person_three:
    return person3
  else:
    return None
      
```

```python
# Shopping List 


class ShoppingList:
    def __init__(self):
        self.products = []

    def number_of_items(self):
        return len(self.products)

    def add(self, product: str, number: int):
        self.products.append((product, number))

    def item(self, n: int):
        return self.products[n - 1][0]

    def amount(self, n: int):
        return self.products[n - 1][1]


    
def total_units(my_list : list):
    unit = 0
    for i in range(1, my_list.number_of_items()+1):
        unit += my_list.amount(i)

    return unit

        
```

```python
#Example of methods in classes

class BankAccount:

    def __init__(self, account_number: str, owner: str, balance: float, annual_interest: float):
        self.account_number = account_number
        self.owner = owner
        self.balance = balance
        self.annual_interest = annual_interest

    # This method adds the annual interest to the balance of the account
    def add_interest(self):
        self.balance += self.balance * self.annual_interest

    # This method "withdraws" money from the account
    # If the withdrawal is successful the method returns True, and False otherwise
    def withdraw(self, amount: float):
        if amount <= self.balance:
            self.balance -= amount
            return True

        return False
```

```python
# Decreasing Counter 

class DecreasingCounter:
    def __init__(self, initial_value: int):
        self.value = initial_value
        self.base_value = initial_value

    def print_value(self):
        print("value:", self.value)

    def decrease(self):
        if self.value >= 1:
            self.value -= 1

    def set_to_zero(self):
        self.value = 0

    def reset_original_value(self):
        self.value = self.base_value
        

counter = DecreasingCounter(55)
counter.decrease()
counter.decrease()
counter.decrease()
counter.decrease()
counter.print_value()
counter.reset_original_value()
counter.print_value()
```

```python
# Example 1: the Rectangle Class

class Rectangle:
    def __init__(self, left_upper: tuple, right_lower: tuple):
        self.left_upper = left_upper
        self.right_lower = right_lower
        self.width = right_lower[0]-left_upper[0]
        self.height = right_lower[1]-left_upper[1]

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return self.width * 2 + self.height * 2

    def move(self, x_change: int, y_change: int):
        corner = self.left_upper
        self.left_upper = (corner[0]+x_change, corner[1]+y_change)
        corner = self.right_lower
        self.right_lower = (corner[0]+x_change, corner[1]+y_change)
```

```python
# Stopwatch

class Stopwatch:
    def __init__(self):
        self.seconds = 0
        self.minutes = 0

    def tick(self):

        if self.seconds >= 59:
                self.seconds = 0
                self.minutes += 1
                
                if self.minutes >= 60:
                    self.minutes = 0
                    
        else:
            self.seconds += 1

    def __str__(self):
        return f'{self.minutes:02}:{self.seconds:02}'

watch = Stopwatch()
for i in range(3600):
    print(watch)
    watch.tick()
        
        
```

```python
# Clock 

class Clock:

    def __init__(self, hour:int, minute:int, second:int):
        self.hour = hour
        self.minute = minute
        self.second = second
        
    def tick(self):
        
        if self.second >= 59:
            self.second = 0
            self.minute += 1
                
            if self.minute >= 60:
                self.minute = 0
                self.hour += 1

                if self.hour >= 24:
                    self.hour = 0
            
        else:
            self.second += 1


    def set(self, set_hour, set_min):
        self.hour = set_hour
        self.minute = set_min
        self.second = 0

    def __str__(self):
        return f'{self.hour:02}:{self.minute:02}:{self.second:02}'
        

clock = Clock(23, 59, 55)
print(clock)
clock.tick()
print(clock)
clock.tick()
print(clock)
clock.tick()
print(clock)
clock.tick()
print(clock)
clock.tick()
print(clock)
clock.tick()
print(clock)
clock.set(12, 5)
print(clock)

```

```python
# Lunch Card

class LunchCard:

    def __init__(self, balance: float):
        self.balance = balance


    def __str__(self):
        return f'The balance is {self.balance:.1f} euros'

    def eat_lunch(self):
        if self.balance >= 2.60:
            self.balance -= 2.60

    def eat_special(self):
        if self.balance >= 4.60:
            self.balance -= 4.60

    def deposit_money(self, money:float):
        if money >= 0:
            self.balance += money
        else:
            raise ValueError("You cannot deposit an amount of money less than zero")


peters_card = LunchCard(20)
graces_card = LunchCard(30)

def main():

    p = 'Peter'
    g = 'Grace'

    peters_card.eat_special()
    graces_card.eat_lunch()

    print(f'{p}: {peters_card}')
    print(f'{g}: {graces_card}')

    peters_card.deposit_money(20)
    graces_card.eat_special()

    print(f'{p}: {peters_card}')
    print(f'{g}: {graces_card}')

    peters_card.eat_lunch()
    peters_card.eat_lunch()
    graces_card.deposit_money(50)

    print(f'{p}: {peters_card}')
    print(f'{g}: {graces_card}')


main()
```

```python
# Series 

class Series:

    def __init__(self, title: str, seasons: int, genres: list):
        self.title = title
        self.seasons = seasons
        self.genres = genres
        self.complete_rating = 0 
        self.count = 0
    
    def rate(self, rating):
        self.complete_rating += rating
        self.count += 1 

    def __str__(self):
        if self.count == 0:
            avg = 0
        else:
            avg = self.complete_rating / self.count

        return (f"{self.title} ({self.seasons} seasons)\n"
                f"genres: {', '.join(self.genres)}\n"
                f"{self.count} ratings, average {avg:.1f} points")

def minimum_grade(rating: float, series_list: list):

    return [series for series in series_list if series.complete_rating >= rating]

def includes_genre(genre:str, series_list: list):

    return [series for series in series_list if genre in series.genres]

s1 = Series("Dexter", 8, ["Crime", "Drama", "Mystery", "Thriller"])
s1.rate(5)

s2 = Series("South Park", 24, ["Animation", "Comedy"])
s2.rate(3)

s3 = Series("Friends", 10, ["Romance", "Comedy"])
s3.rate(2)

series_list = [s1, s2, s3]

print("a minimum grade of 4.5:")
for series in minimum_grade(4.5, series_list):
    print(series.title)

print("genre Comedy:")
for series in includes_genre("Comedy", series_list):
    print(series.title)
```
