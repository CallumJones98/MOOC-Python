```python
# The Fastest Car 

class Car:
    def __init__(self, make: str, top_speed: int):
        self.make = make
        self.top_speed = top_speed

    def __str__(self):
        return f"Car (make: {self.make}, top speed: {self.top_speed})"


def fastest_car(cars: list):
    max_speed = 0
    chosen_car = None

    for car in cars:
        if car.top_speed >= max_speed:
            chosen_car = car.make
            max_speed = car.top_speed
        else:
            continue

    return chosen_car

car1 = Car("Saab", 195)
car2 = Car("Lada", 110)
car3 = Car("Ferrari", 280)
car4 = Car("Trabant", 85)
cars = [car1, car2, car3, car4]
print(fastest_car(cars))
```

```python
# Alternative version using lambda function

def fastest_car(cars: list):
    
    if not cars:
        return None
    
    fastest = max(cars, key=lambda car: car.top_speed)
    
    return fastest.make
```

```python
# Passing Submissions

class ExamSubmission:
    def __init__(self, examinee: str, points: int):
        self.examinee = examinee
        self.points = points

    def __str__(self):
        return f'ExamSubmission (examinee: {self.examinee}, points: {self.points})'

def passed(submissions: list, lowest_passing: int):

    return [student for student in submissions if student.points >= lowest_passing]

s1 = ExamSubmission("Peter", 12)
s2 = ExamSubmission("Pippa", 19)
s3 = ExamSubmission("Paul", 15)
s4 = ExamSubmission("Phoebe", 9)
s5 = ExamSubmission("Persephone", 17)

passes = passed([s1, s2, s3, s4, s5], 15)

for passing in passes:
        print(passing)
```
